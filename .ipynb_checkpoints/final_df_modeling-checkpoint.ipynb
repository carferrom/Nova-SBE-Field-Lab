To evaluate the impact of programming activity in a region we must first define what is high programming activity? 
Based on our dataset analysis of stackverflow activity, we can analyze 4 different criteria (Question, Answer, Upvote and Downvote counts). For a better analysis we are defining a metric **"Programing Activity"** to which we assign weights to the 4 criteria according to the following assumptions:

**NOTE CHAT-GPT GENERATED LETS CHANGE THIS**

**Question Count (40%):** Questions are the core of a Q&A platform and drive the community's engagement. A higher volume of questions indicates a thriving environment where knowledge seeking is prevalent. This is often the starting point of community interactions, hence a higher weight.

**Answer Count (30%):** Answers are direct indicators of community responsiveness. While they are crucial, they are a step that follows questions, and their frequency is dependent on the questions asked, hence a slightly lower weight.

**Upvote Count (20%):** Upvotes are a measure of community agreement and the value found in questions and answers. They indicate the usefulness and quality of contributions, but as they are secondary to the existence of content, they get a lower weight.

**Downvote Count (10%):** Downvotes show community engagement and content curation by signaling less accurate or less useful content. However, because they indicate negative feedback and are less central to knowledge exchange, they are weighted the least.

For a better understanding of this metric it will be normalized to fall within the range of 0 to 1, where 1 represents the highest level of programming activity detected in the dataset and 0 the lowest.

```python
# Define the weights for each criteria
weights = {
    'questioncount': 0.4,
    'answercount': 0.3,
    'upvotecount': 0.2,
    'downvotecount': 0.1
}

# Calculate the Programming Activity metric
FINAL_no_TR['Programming Activity'] = (FINAL_no_TR['questioncount'] * weights['questioncount'] +
                                        FINAL_no_TR['answercount'] * weights['answercount'] +
                                        FINAL_no_TR['upvotecount'] * weights['upvotecount'] +
                                        FINAL_no_TR['downvotecount'] * weights['downvotecount']) / 4


max_activity = FINAL_no_TR['Programming Activity'].max()

# Normalize the 'Programming Activity' to a 0-1 scale
FINAL_no_TR['Programming Activity Index'] = FINAL_no_TR['Programming Activity'] / max_activity

# Drop the original columns
columns_to_drop = ['questioncount', 'answercount', 'upvotecount', 'downvotecount', 'commentcount','Programming Activity']
FINAL_no_TR.drop(columns=columns_to_drop, inplace=True)

# Save the final dataframe to an Excel file
FINAL_no_TR.to_excel("Programming Index.xlsx", index=False)

FINAL_no_TR
```

```python
max_activity = FINAL_no_TR['Programming Activity Index'].max()
min_activity = FINAL_no_TR['Programming Activity Index'].min()

print(f"Maximum Programming Activity: {max_activity}")
print(f"Minimum Programming Activity: {min_activity}")
```

```python
# Aggregating numeric data by nuts3_code
aggregated_numeric_data = FINAL_no_TR.groupby('nuts3_code').mean().reset_index()

# Selecting non-aggregated columns
non_aggregated_data = FINAL_no_TR[['nuts3_code', 'country', 'nuts3_name']].drop_duplicates()

# Merging aggregated and non-aggregated data
aggregated_data = pd.merge(non_aggregated_data, aggregated_numeric_data, on='nuts3_code', how='left')

# Reordering the columns to place 'country' and 'nuts3_name' first
column_order = ['country', 'nuts3_name', 'nuts3_code'] + [col for col in aggregated_data.columns if col not in ['country', 'nuts3_name', 'nuts3_code']]
aggregated_data = aggregated_data[column_order]
aggregated_data
```

```python
# Range of possible clusters
cluster_range = range(1, 11)

# Calculate inertia for each number of clusters
inertia_values = []
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(aggregated_data[['Programming Activity Index']])
    inertia_values.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o')
plt.title('Elbow Method For Optimal Cluster Number')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

The elbow method is used to choose an appropiate number of clusters, which visually would be at the point where the curve begins to decrease more slowly, which would be around 3 clusters. After this point, the curve starts decreasing at a much lower rate, meaning that using more clusters doesnt neccesarily provide any significantly better fitting of the data.

```python
# Optimal Clusters
optimal_clusters = 3 

# Applying K-Means Clustering on aggregated data
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
aggregated_data['Cluster'] = kmeans.fit_predict(aggregated_data[['Programming Activity Index']])

# Plotting the clusters against aggregated GDP
plt.figure(figsize=(10, 8))
scatter = plt.scatter(aggregated_data['Programming Activity Index'], aggregated_data['GDP (MIO_EUR)'],
                      c=aggregated_data['Cluster'], cmap='viridis', alpha=0.8, edgecolor='k')

# Create a color bar with discrete labels
cbar = plt.colorbar(scatter, ticks=np.arange(0, optimal_clusters))
cbar.set_label('Cluster')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add labels and title
plt.xlabel('Programming Activity Index', fontsize=14)
plt.ylabel('GDP in Million EUR', fontsize=14)
plt.title('Clusters of NUTS3 Regions by Programming Activity and GDP (Aggregated)', fontsize=16)

# Adjust layout to prevent clipping of ylabel
plt.tight_layout()

# Show the plot
plt.show()
```
