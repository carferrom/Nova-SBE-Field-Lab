### Explain Programming Activity

In the group part we only looked at one cross-section, the year 2016, and explained the programming activity using the Lasso Regression model. Now we want to look at the whole dataset (panel) and also explore if we can build ML models that can explain programming activity even better than the linear models used in the group part. For this we are going to look at the following models: Random Forest Regressor, Gradien Boosting Machine and Neural Networks.

**1) Defining the variables** 

```python
# Define a dataset with the necessary variables
dataset = df[['year', 'nuts3_name', 'activity', 'GDP_per_cap', 'POP (THS)', 'EMP (THS)']]

# 1. Region-Specific Dummies
region_dummies = pd.get_dummies(dataset['nuts3_name'], prefix='region')
dataset = pd.concat([dataset, region_dummies], axis=1)

# 2. Time Trends
dataset['time_trend'] = dataset['year'] - dataset['year'].min()

dataset.head()
```

Now we split the dataset into train and test set using **random sampling within regions.** This randomly samples a proportion of data from each region to create the hold-out set. This method ensures that the validation set is representative of the entire dataset across all regions.

```python
# Define the proportion of data to be used as the hold-out set
hold_out_proportion = 0.2  

# Initialize an empty DataFrame for the hold-out set
hold_out_df = pd.DataFrame()

# Iterate through each region and randomly select a proportion of data
for region in dataset['nuts3_name'].unique():
    region_data = dataset[dataset['nuts3_name'] == region]
    hold_out_region = region_data.sample(frac=hold_out_proportion, random_state=42)
    hold_out_df = pd.concat([hold_out_df, hold_out_region])

# Create the training set by dropping the hold-out set from the original dataset
train_df = dataset.drop(hold_out_df.index)

# Now, train_df can be used for training and hold_out_df for validation
X_train = train_df.drop(['activity', 'nuts3_name', 'year'], axis=1)   
y_train = train_df['activity']  

X_hold_out = hold_out_df.drop(['activity', 'nuts3_name', 'year'], axis=1)  
y_hold_out = hold_out_df['activity'] 
```

```python
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras_tuner import RandomSearch


# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_hold_out_scaled = scaler.transform(X_hold_out)

def build_model(hp):
    model = Sequential()
    model.add(Dense(hp.Int('input_units', min_value=32, max_value=256, step=32),
                    activation='relu', input_shape=(X_train_scaled.shape[1],)))

    for i in range(hp.Int('n_layers', 1, 4)):
        model.add(Dense(hp.Int(f'dense_{i}_units', min_value=16, max_value=128, step=16),
                        activation='relu'))

    model.add(Dense(1, activation='linear'))
    model.compile(optimizer=tf.keras.optimizers.Adam(hp.Float('learning_rate', 1e-4, 1e-2, sampling='log')),
                  loss='mean_squared_error')
    return model

# Set up Keras Tuner
tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,  # Set the number of trials (model configurations to test)
    executions_per_trial=1,
    directory='my_dir',
    project_name='keras_tuning'
)

tuner.search(X_train_scaled, y_train, epochs=10, validation_data=(X_hold_out_scaled, y_hold_out))

# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Fit the best model
best_model.fit(X_train_scaled, y_train, epochs=50, batch_size=32)  

# Predictions and evaluation
y_pred = best_model.predict(X_hold_out_scaled)
mse = mean_squared_error(y_hold_out, y_pred)
r2 = r2_score(y_hold_out, y_pred)

print(f'Mean Squared Error on Test Set: {mse}')
print(f'R-squared on Test Set: {r2}')
```

**3) Backing out the Residuals (epsilon)**

We are going to back out the residuals for each year and each NUTS3 region in order to be able to use it in our panel data regression.

```python
# Calculate residuals for each year using the best model
residuals = {}

for year in range(2008,2021):
    data_year = dataset[dataset['year'] == year]
    X = data_year[['GDP_per_cap', 'EMP (THS)', 'POP (THS)']]
    y = data_year['activity']

    model = best_model
    model.fit(X, y)
    y_pred = model.predict(X)
    residuals[year] = y - y_pred

# Combine residuals into the main dataset
dataset['activity_eps'] = dataset.apply(lambda row: residuals[row['year']][row.name], axis=1)

dataset.head()
```

```python
# Descriptive Statistics of Residuals
residual_stats = dataset['activity_eps'].describe()
print("Descriptive Statistics of Residuals:\n", residual_stats)

# Histogram of Residuals
plt.figure(figsize=(10, 6))
sns.histplot(dataset['activity_eps'], kde=True)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Residuals vs Predicted Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=(dataset['activity'] - dataset['activity_eps']), y=dataset['activity_eps'])
plt.axhline(0, color='red', linestyle='--')  # Adds a reference line at zero
plt.title('Residuals vs. Predicted Values')
plt.xlabel('Predicted Programming Activity')
plt.ylabel('Residuals')
plt.show()
```

**4) Predicting Future Growth of GDP**

In order to regress future growth of GDP we first need to calculate the growth variable, because it will be the independent target variable that we want to predict. We will choose a 3 year gap to calculate growth of GDP. 

```python
# Calculate 3 year GDP growth
def calculate_3y_growth(data):
    # Sorting the data to ensure it's in the right order
    data = data.sort_values(['nuts3_name', 'year'])

    # Using groupby to iterate over each region separately
    for region, group in data.groupby('nuts3_name'):
        # Using shift to get the GDP value from 3 years ago
        data.loc[group.index, 'GDP_3y_growth'] = group['GDP_per_cap'] - group['GDP_per_cap'].shift(3)

    return data

# Apply the function to the dataset
dataset_with_growth = calculate_3y_growth(dataset)

# Display the modified dataset (optional)
dataset_with_growth.head(10)
```

Now we regress the growth variable one time with and without using the residuals to compare the results.

**4a) Without Residuals**

```python
from linearmodels import PanelOLS, RandomEffects, PooledOLS
import statsmodels.api as sm

# Set MultiIndex
dataset_regression = dataset_with_growth[dataset_with_growth.year >= 2011]
dataset_regression = dataset_regression.set_index(['nuts3_name', 'year'])

exog = sm.tools.tools.add_constant(dataset_regression[['activity','GDP_per_cap','POP (THS)']])
endog = dataset_regression['GDP_3y_growth']

# Pooled OLS
mod_pooled_ols = PooledOLS(endog, exog)
pooled_ols_res = mod_pooled_ols.fit()

# Output results
print("Pooled OLS Results:\n", pooled_ols_res)
```
